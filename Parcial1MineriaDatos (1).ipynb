```python
G1 = list(df['G1'])
G2 = list(df['G2'])
G3 = list(df['G3'])
nota_final=list()
for i in range (len(G1)):
    valor = (G1[i]*0.30)+(G2[i]*0.30)+(G3[i]*0.40)
    nota_final.append(valor)
df1 =pd.DataFrame( {'Nota 1 ' : G1,
          'Nota 2' : G2, 'Nota 3':G3, 'Nota final' : nota_final})
df1.describe()
```

```python
df1.median()
```
